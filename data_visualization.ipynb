```python
# 研究变量和click的关系
def plot_click_rate(data, xcoordinate, ycoordinate):
    data_count = data[[xcoordinate, ycoordinate]].groupby(xcoordinate).count().reset_index()
    data_sum = data[[xcoordinate, ycoordinate]].groupby(xcoordinate).sum().reset_index()
    data_sum['click_rate'] = data_sum[ycoordinate] / data_count[ycoordinate]
    data_final = data_sum.drop([ycoordinate], axis = 1)
    del data_count, data_sum
    sns.barplot(data = data_final, x = xcoordinate, y = 'click_rate', palette = 'hls', order = data_final.sort_values('click_rate')[xcoordinate])
    temp_path = "C:\\Users\\lenovo\\Downloads\\Thesis\\code1\\" + xcoordinate + "_rate.png"
    plt.savefig(temp_path, dpi = 200, bbox_inches = 'tight')
    plt.show()
```

```python
for col in col_graph[1:]:
    #print(col)
    plot_click_rate(data_drop, col, 'label')
```
